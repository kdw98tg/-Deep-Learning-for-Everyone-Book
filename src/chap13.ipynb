- 지금까지 테스트셋을 만들지 않고 했음. 근데 매번 정확도를 계산할 수 있었음. 
- 모든 샘플을 그대로 테스트에 활용했음. 그래서 주어진 테스트 데이터에 잘 맞는 모델을 만든거고.

```python
from tensorflow.keras.models import Sequential, load_model
from tensorflow.keras.layers import Dense
from sklearn.model_selection import train_test_split
# 세이킷 런 라이브러리는 머신러닝을 실행할 때 필요한 전반적인 것들이 담긴 머신러닝의 필수 라이브러리임

import pandas as pd
```

```python
df = pd.read_csv('../data/sonar3.csv', header = None)
X = df.iloc[:,0:60]
y= df.iloc[:,60]
len(X)
```

```python
# 학습 데이터셋을 70% 테스트 셋을 30%로 설정한 예
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, shuffle = True)
```

```python
model = Sequential()
model.add(Dense(24, input_dim = 60, activation='relu'))
model.add(Dense(10, activation='relu'))
model.add(Dense(1, activation='sigmoid'))

model.compile(loss='binary_crossentropy', optimizer = 'adam', metrics=['accuracy'])

history = model.fit(X_train, y_train, epochs = 200, batch_size=10)
```

```python
score = model.evaluate(X_test, y_test)# 훈련되지 않은 값들을 넣어서 출력의 리스트를 반환 [loss, accurracy] 위에서 설정한것과 같음
print('test accuracy :', score[1])
```

## K겹 교차 검증
- 학습 데이터를 70%, 테스트 데이터셋을 30%로 하면 괜찮아 보이지만, 데이터의 30%나 학습하는데 사용할 수 없다는것임
- 그래서 k겹 교차 검증을 사용함
- 데이터셋을 여러개로 나누어 하나씩 테스트셋으로 사용하고 나머지를 모두 합해서 학습셋으로 사용하는 방법임
- 즉, 5겹 교차 검증이란 데이터셋을 5등분한것(4등분은 학습, 1등분은 테스트 -> 테스트셋을 번갈아가며 5번 반복)


```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from sklearn.model_selection import KFold
from sklearn.metrics import accuracy_score
import pandas as pd
```

```python
df = pd.read_csv('../data/sonar3.csv', header = None)

X = df.iloc[:, 0:60]
y = df.iloc[:,60]
```

```python
# 몇겹
k = 5

# KFold 함수를 불러옴. 분할하기 전에 샘플이 치우치지 않도록 섞음
kfold = KFold(n_splits=k, shuffle = True)

# 정확도가 채워질 빈 리스트 (k=5 니까 5개 채워질거임)
acc_score = []

def model_fn():
    model = Sequential()
    model.add(Dense(units = 24, input_dim = 60, activation = 'relu'))
    model.add(Dense(units = 10, activation = 'relu'))
    model.add(Dense(units = 1, activation = 'sigmoid'))
    return model

for train_index, test_index in kfold.split(X):
    X_train, X_test = X.iloc[train_index,:], X.iloc[test_index,:]
    y_train, y_test = y.iloc[train_index], y.iloc[test_index]

    model = model_fn()
    model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
    history=model.fit(X_train, y_train, epochs=200, batch_size=10, verbose=0)

    accuracy = model.evaluate(X_test, y_test)[1]
    acc_score.append(accuracy)

avg_acc_score = sum(acc_score)/k

print("정확도: ",acc_score)
print("정확도 평균: ", avg_acc_score)
```
